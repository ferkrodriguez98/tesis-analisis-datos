```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
csv_file_path = 'data/data_all_cleaned.csv'
data_raw = pd.read_csv(csv_file_path)

data = data_raw.copy()

print('Cantidad de participantes: ', len(data))
data.head()
```

```python
sns.set_style("whitegrid")

# GrÃ¡fico 1.1: DistribuciÃ³n de categorÃ­a de estudios
plt.figure(figsize=(12, 3))
plt.subplot(1, 2, 1)
sns.countplot(data=data, x='degree_category', order=data['degree_category'].value_counts().index)
plt.title('DistribuciÃ³n de categorÃ­a de estudios')
plt.xlabel('CategorÃ­a de estudios')
plt.ylabel('Frecuencia')
plt.xticks(rotation=60)

# GrÃ¡fico 1.2: DistribuciÃ³n de categorÃ­a de estudios por gÃ©nero
plt.subplot(1, 2, 2)
sns.countplot(data=data, x='degree_category', hue='gender', order=data['degree_category'].value_counts().index)
plt.title('DistribuciÃ³n de categorÃ­a de estudios por gÃ©nero')
plt.xlabel('CategorÃ­a de estudios')
plt.ylabel('Frecuencia')
plt.xticks(rotation=60)

plt.tight_layout()
plt.show()
```

```python
# GrÃ¡fico 2: DistribuciÃ³n de paÃ­s de nacimiento
plt.figure(figsize=(6, 3))
sns.countplot(data=data, x='birth_location_country_clean', order=data['birth_location_country_clean'].value_counts().index)
plt.title('DistribuciÃ³n de paÃ­s de nacimiento')
plt.xlabel('PaÃ­s')
plt.ylabel('Frecuencia')
plt.xticks(rotation=90)
plt.show()
```

```python
# GrÃ¡fico 3.1: DistribuciÃ³n de nivel de felicidad al momento del test
plt.figure(figsize=(10, 3))
plt.subplot(1, 2, 1)
sns.countplot(data=data, x='happiness_level')
plt.title('DistribuciÃ³n de nivel de felicidad')
plt.xlabel('Nivel de felicidad')
plt.ylabel('Frecuencia')

# GrÃ¡fico 3.2: DistribuciÃ³n de nivel de felicidad al momento del test por gÃ©nero
plt.subplot(1, 2, 2)
sns.countplot(data=data, x='happiness_level', hue='gender')
plt.title('DistribuciÃ³n de nivel de felicidad por gÃ©nero')
plt.xlabel('Nivel de felicidad')
plt.ylabel('Frecuencia')
plt.xticks(rotation=60)

plt.show()
```

```python
scores_and_gender = data[['gender', 'openness_score', 'conscientiousness_score', 'extraversion_score', 'agreeableness_score', 'neuroticism_score']]

scores_and_gender['is_female'] = (scores_and_gender['gender'] == 'F').astype(int)
scores_and_gender['is_male'] = (scores_and_gender['gender'] == 'M').astype(int)

scores_and_gender = scores_and_gender.drop('gender', axis=1)

correlation_matrix = scores_and_gender.corr()

correlation_with_gender = correlation_matrix.loc[['is_female', 'is_male'], ['openness_score', 'conscientiousness_score', 'extraversion_score', 'agreeableness_score', 'neuroticism_score']]

mean_scores = data[data['gender'].isin(['F', 'M'])].groupby('gender')[['openness_score', 'conscientiousness_score', 'extraversion_score', 'agreeableness_score', 'neuroticism_score']].mean()

mean_scores = mean_scores.T

fig, ax = plt.subplots(1, 2, figsize=(12, 5), gridspec_kw={'width_ratios': [1, 2]})

# GrÃ¡fico 11.1: Scores promedio por gÃ©nero
sns.heatmap(mean_scores, annot=True, fmt=".2f", cmap='coolwarm', cbar=False, ax=ax[0], linewidths=.5)

ax[0].set_title('Score promedio por gÃ©nero')

# GrÃ¡fico 11.2: CorrelaciÃ³n entre rasgos de personalidad y gÃ©nero
sns.heatmap(correlation_with_gender.T, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5, cbar_kws={'orientation': 'horizontal'}, ax=ax[1])
ax[1].set_title('CorrelaciÃ³n entre gÃ©nero y rasgo de personalidad')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
traits = ['openness_score', 'conscientiousness_score', 'extraversion_score', 'agreeableness_score', 'neuroticism_score']
trait_names = ['Openness', 'Conscientiousness', 'Extraversion', 'Agreeableness', 'Neuroticism']

# GrÃ¡fico: Violin plots de cada OCEAN trait por gÃ©nero
plt.figure(figsize=(20, 4))

for i, (trait, trait_name) in enumerate(zip(traits, trait_names)):
    plt.subplot(1, 5, i+1)
    sns.violinplot(data=data[data['gender'].isin(['M', 'F'])], x='gender', y=trait, split=True, palette=['lightblue', 'pink'], linewidth=1, hue='gender', inner='quartile')
    plt.title(trait_name)
    plt.xlabel('GÃ©nero')
    plt.ylabel('Score')
    if i > 0:
        plt.ylabel('')

plt.tight_layout()
plt.show()

# GrÃ¡fico: Violin plots de cada OCEAN trait por happiness level
plt.figure(figsize=(20, 4))

for i, (trait, trait_name) in enumerate(zip(traits, trait_names)):
    plt.subplot(1, 5, i+1)
    sns.violinplot(data=data, x='happiness_level', y=trait, linewidth=1, inner='quartile', hue='happiness_level', palette=sns.cubehelix_palette(5, start=.1, gamma=0.3), legend=False)
    plt.title(trait_name)
    plt.xlabel('GÃ©nero')
    plt.ylabel('Score')
    if i > 0:
        plt.ylabel('')

plt.tight_layout()
plt.show()
```
